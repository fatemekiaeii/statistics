```python
k=df.plot.scatter(x='a', y='b',color='red',label='group1')
df.plot.scatter(x='c',y='d',color='blue',label='group2',ax=k)
```

# pie plot

```python
l=['python','java','c++']
s=[200, 150,100]
c=["green",'red','yellow']
```

```python
plt.pie(s,labels=l,colors=c);
```

```python
plt.pie(s, labels=l, colors=c,  autopct='%1.1f%%');
```

```python
plt.pie(s, labels=l, colors=c,startangle=180);
```

```python
plt.pie(s, labels=l, colors=c,explode=(0.1,0,0),shadow=True,autopct='%1.1f%%')
```
